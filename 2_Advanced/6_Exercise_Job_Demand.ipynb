```python
# Importing Libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("Lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no']=pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot['month_no']
```

```python
df_US_pivot.info()
print(df_US_pivot.head())
```

```python
# 1️⃣ Remove the unnecessary index columns first
df_US_pivot.drop(columns=['level_0', 'index'], inplace=True)

# 2️⃣ Create numeric month order
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

# 3️⃣ Sort by numeric month
df_US_pivot.sort_values('month_no', inplace=True)

# 4️⃣ Set the month name as index
df_US_pivot.set_index('job_posted_month', inplace=True)

# 5️⃣ Drop the helper column
df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_US_pivot[top_3].plot(kind='line')
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Job Demand Over Months for Top 3 Job Titles in the US')
plt.xlabel('2023')
plt.ylabel('Number of Job Postings')
plt.legend(title='Job Titles')
plt.show()
```
