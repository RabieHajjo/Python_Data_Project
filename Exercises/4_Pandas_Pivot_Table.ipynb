```python
# Importing Libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("Lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')
```

```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.groupby('job_title_short').size()
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', columns='job_country', aggfunc='median')
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_title_short',
    columns='job_country',
    aggfunc='median'
)

df_top_countries = df_job_country_salary[top_countries]

df_top_countries
```

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst']

df_top_countries = df_job_country_salary.loc[top_countries, job_titles]

df_top_countries.plot(
    kind='bar',
    figsize=(10,6),
    ylabel='Median Salary (USD)',
    title='Median Salary by Country and Job Title'
)

```
